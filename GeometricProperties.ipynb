```python
# Loading Libraries

import numpy as np
import cv2
import matplotlib.pyplot as plt
from skimage.measure import regionprops, label
from google.colab.patches import cv2_imshow
```

```python
# Loading Image

img_path = '7UynE.jpg'
img = cv2.imread(img_path)
grey_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

cv2_imshow(img)
```

```python
# Binary Image Generation

_, binary_image = cv2.threshold(grey_img, 127, 255, cv2.THRESH_BINARY)
```

```python
labeled_image = label(binary_image)
```

```python
regions = regionprops(labeled_image)
```

```python
# Display geometric properties for each region
for idx, region in enumerate(regions, start=1):
    print(f"Region {idx}:")
    print(f"  Area: {region.area}")
    print(f"  Perimeter: {region.perimeter}")
    print(f"  Centroid: {region.centroid}")
    print(f"  Bounding Box: {region.bbox}")
    print(f"  Eccentricity: {region.eccentricity}")
    print(f"  Orientation: {region.orientation}")
    print("-" * 40)

# Display the original binary image
plt.imshow(binary_image, cmap='gray')
plt.title('Binary Image')
plt.axis('off')
plt.show()
```
