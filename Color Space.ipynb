# Color Space

```python
## Loading Libraries

import numpy as np
import cv2
from google.colab.patches import cv2_imshow
```

```python
##  Loading Image

img = cv2.imread('image.png')
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

cv2_imshow(img)
cv2_imshow(img_gray)
```

## Convert RGB to Grayscale

```python
img_BGR_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

cv2_imshow(img_BGR_gray)
```

## Convert RGB to HSV

```python
hsv_image = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

cv2_imshow(hsv_image)
```

## Convert RGB to LAB

```python
lab_image = cv2.cvtColor(img, cv2.COLOR_BGR2LAB)

cv2_imshow(lab_image)
```

## Convert RGB to YCrCb

```python
ycrcb_image = cv2.cvtColor(img, cv2.COLOR_BGR2YCrCb)

cv2_imshow(ycrcb_image)
```

## Combining Color Spaces for Analysis

```python
h, s, v = cv2.split(hsv_image)

cv2.imwrite('hue.jpg', h)
cv2.imwrite('saturation.jpg', s)
cv2.imwrite('value.jpg', v)
```

## Color Space Thresholding Example

```python
hsv_img = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

lower_bound = np.array([0, 120, 70])
upper_bound = np.array([100, 255, 255])

mask = cv2.inRange(hsv_img, lower_bound, upper_bound)

out_img = cv2.bitwise_and(img, hsv_image, mask=mask)

cv2_imshow(mask)
cv2_imshow(out_img)
```

## Thresholding in HSV

```python
lab_img = cv2.cvtColor(img, cv2.COLOR_BGR2LAB)

lower_bound = np.array([0, 120, 70])
upper_bound = np.array([100, 150, 150])

mask = cv2.inRange(lab_img, lower_bound, upper_bound)

out_img = cv2.bitwise_and(img, lab_img, mask=mask)

cv2_imshow(mask)
cv2_imshow(out_img)
```
