# Image Filtering Techniques: Sharpening and Smoothing

```python
## Loading Libraries

import numpy as np
import cv2
from google.colab.patches import cv2_imshow
```

```python
## Loading Image

img = cv2.imread('image.png')
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

cv2_imshow(img)
cv2_imshow(img_gray)
```

## Smoothing Filter

## Gaussian Smoothing / Blur

```python
window_size = (15, 5) # shall be odd
sigma=0
gaussian_blur_img = cv2.GaussianBlur(img, window_size, sigma=sigma)

cv2_imshow(gaussian_blur_img)
```

## Median Blur


```python
kernel_size = 5

median_blur_img = cv2.medianBlur(img, kernel_size)

cv2_imshow(median_blur_img)
```

## Bilateral Filter

```python
diameter = 9
sigma_color, sigma_space = 75, 75

bil_img = cv2.bilateralFilter(img, diameter, sigma_color, sigma_space)

cv2_imshow(bilateral_filter_img)
```

## Non local means denoising (More denoising techniques than smoothing)

```python
nlm_denoise_img = cv2.fastNlMeansDenoisingColored(img, None, h=10, templateWindowSize=7, searchWindowSize=31)

cv2_imshow(nlm_denoise_img)
```

## Average Filter / Box Filter

```python
kernel_size = (5, 5)
avg_img = cv2.blur(img, kernel_size)

cv2_imshow(avg_img)
```

## Sharpening Filters

## Custom Kernel

```python
kernel = np.array([[0, -1, 0],
                   [-1, 5, 1],
                   [0, -1, 0]])

sharp_img = cv2.filter2D(img, -1, kernel)

cv2_imshow(sharp_img)
```

## Using Unsharp Masking

```python
blur = cv2.GaussianBlur(img, (9, 9), 10)

sharp_img = cv2.addWeighted(img, 1.5, blur, -0.5, -0)

cv2_imshow(sharp_img)
```

## High pass filter

```python
kernel = np.array([[-1, -1, -1],
                   [-1, 8, -1],
                   [-1, -1, -1]])

high_pass = cv2.filter2D(img, -1, kernel)

cv2_imshow(high_pass)
```
