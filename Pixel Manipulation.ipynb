# Pixel Manipulation

```python
## Loading Libraries

import numpy as np
import cv2
from google.colab.patches import cv2_imshow
```

```python
## Loading Image

img = cv2.imread('image.png')
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

cv2_imshow(img)
cv2_imshow(img_gray)
```

## Contrast and Brightness Correction: Linear Transformation

```python
alpha = 1.5 # Contrast, range: 1.0 to 3.0
beta = 20 # Brightness, range: 0 to 100

adjusted_img = cv2.convertScaleAbs(img, alpha=alpha, beta=beta)

cv2_imshow(adjusted_img)
```

## Contract Adjustment: Gamma Correction

```python
def adjust_gamma(img, gamma=1.0):
  inv_gamma=1.0/gamma
  table = np.array([((i / 255.0) ** inv_gamma) * 255 for i in np.arange(0, 256)]).astype("uint8")
  return cv2.LUT(img, table)

gamma_img = adjust_gamma(img, gamma=1.5)
cv2_imshow(gamma_img)
```

## Negative Transformation

We can also add darkness or brightness such as img-x. where x is amount of darkness/brightness. we require.

Pixel processing: Low contrast (img/2), High Contrast (img*2), Gray (0.3*imgRchannel + 0.6*imgGchannel + 0.1*imgBchannel)

```python
# Method 1

neg_img = 255-img

cv2_imshow(neg_img)
```

```python
## Method 2

neg_img = cv2.bitwise_not(img)
cv2_imshow(neg_img)
```
