# Morphological Operations

```python
## Loading Libraries

import numpy as np
import cv2
from google.colab.patches import cv2_imshow
```

```python
##  Loading Image

img = cv2.imread('image.png')
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

cv2_imshow(img)
cv2_imshow(img_gray)
```

## Erosion

```python
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

_, bin_img = cv2.threshold(img_gray, 127, 255, cv2.THRESH_BINARY)

kernel = np.ones((3,3), np.uint8)

ero_img = cv2.erode(bin_img, kernel, iterations=1)

cv2_imshow(ero_img)
```

```python
rect_kernal = cv2.getStructuringElement(cv2.MORPH_RECT, (5, 5))

ellipse_kernal = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (5, 5))

cross_kernel = cv2.getStructuringElement(cv2.MORPH_CROSS, (5, 5))

ero_img_rect = cv2.erode(bin_img, rect_kernal, iterations=1)
ero_img_ellipse = cv2.erode(bin_img, ellipse_kernal, iterations=1)
ero_img_cross = cv2.erode(bin_img, cross_kernel, iterations=1)

cv2_imshow(ero_img_rect)
cv2_imshow(ero_img_ellipse)
cv2_imshow(ero_img_cross)
```

## Dilation

```python
kernel = np.ones((3, 3), np.uint8)

dilated_img = cv2.dilate(bin_img, kernel, iterations=1)

cv2_imshow(dilated_img)
```

```python
rect_kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (5, 5))
ellipse_kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (5, 5))
cross_kernel = cv2.getStructuringElement(cv2.MORPH_CROSS, (5, 5))

dilated_img_rect = cv2.dilate(bin_img, rect_kernel, iterations=1)
dilated_img_ellipse = cv2.dilate(bin_img, ellipse_kernel, iterations=1)
dilated_img_cross = cv2.dilate(bin_img, cross_kernel, iterations=1)

cv2_imshow(dilated_img_rect)
cv2_imshow(dilated_img_ellipse)
cv2_imshow(dilated_img_cross)
```

## Opening

```python
kernel = np.ones((5, 5), np.uint8)

opened_image = cv2.morphologyEx(bin_img, cv2.MORPH_OPEN, kernel)

cv2_imshow(opened_image)
```

## Closing

```python
kernel = np.ones((5, 5), np.uint8)

closed_image = cv2.morphologyEx(bin_img, cv2.MORPH_CLOSE, kernel)

cv2_imshow(closed_image)
```
