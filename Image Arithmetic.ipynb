## Image Arithmetic

```python
## Loading Libraries

import numpy as np
import cv2
from google.colab.patches import cv2_imshow
```

```python
##  Loading Image

img1 = cv2.imread('image.png')
img_gray1 = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)

img2 = cv2.imread('image.png')
img_gray2 = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)

cv2_imshow(img1)
cv2_imshow(img_gray1)
cv2_imshow(img2)
cv2_imshow(img_gray2)
```

## Addition of Images: Simple Addition

```python
addedIMG = cv2.add(img2, img1)

cv2_imshow(addedIMG)
```

## Weighted Addition / Blending

```python
alpha, beta, gamma = 0.5, 0.5, 0

blendedIMG = cv2.addWeighted(img1, alpha, img2, beta, gamma)

cv2_imshow(blendedIMG)
```

## Subsctraction of Image

```python
substractedIMG = cv2.subtract(img2, img1)

cv2_imshow(substractedIMG)
```

## Multiplication of Images: Elementwise

```python
multipliedIMG = cv2.multiply(img2, img1)

cv2_imshow(multipliedIMG)
```

## Multiplication of Images: Scalar

```python
multiplyIMG = cv2.multiply(img1, np.array([1.5, 1.5, 1]))

cv2_imshow(multiplyIMG)
```

## Division of Images

```python
divIMG = cv2.divide(img2, img1)

cv2_imshow(divIMG)
```

## Bitwise Operation

### Bitwise AND

```python
bitwise_and = cv2.bitwise_and(img1, img2)

cv2_imshow(bitwise_and)
```

## Bitwise Or

```python
bitwise_or = cv2.bitwise_or(img1, img2)

cv2_imshow(bitwise_or)
```

### Bitwise XOR

```python

bitwiseXOR = cv2.bitwise_xor(img1, img2)

cv2_imshow(bitwiseXOR)
```

### Bitwise NOT

```python
bitwise_not = cv2.bitwise_not(img1)

cv2_imshow(bitwise_not)
```
