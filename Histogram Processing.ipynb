# Histogram Processing

```python
## Loading Libraries

import numpy as np
import cv2
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
```

```python
##  Loading Image

img = cv2.imread('img_gray.jpg')
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

cv2_imshow(img)
cv2_imshow(img_gray)
```

```python
pixels = img_gray.ravel()

# Plot the histogram
plt.hist(pixels, bins=25, range=[0, 256], color='gray', edgecolor='black')
plt.title('Histogram of Grayscale Image')
plt.xlabel('Pixel Intensity')
plt.ylabel('Frequency')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

## Histogram Equalization

```python
equalized_img = cv2.equalizeHist(img_gray)

cv2_imshow(equalized_img)
```

## Histogram Equalization for Color Image

```python
# Convert the image to YUV color space
yuv_img = cv2.cvtColor(img, cv2.COLOR_BGR2YUV)

yuv_img[:, :, 0] = cv2.equalizeHist(yuv_img[:, :, 0])

equalized_img = cv2.cvtColor(yuv_img, cv2.COLOR_YUV2BGR)

cv2_imshow(equalized_img)
```

## CLAHE - Contrast Limited Adaptive Histogram Equalization

```python
clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8, 8))
clahe_img = clahe.apply(img_gray)

cv2_imshow(clahe_img)
```
