# Image Normalization

```python
# Loading Libraries

import numpy as np
import cv2
from google.colab.patches import cv2_imshow
```

```python
# Loading Image

img = cv2.imread('image.png')
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

cv2_imshow(img)
cv2_imshow(img_gray)
```

## Min max normalization

```python
normalized_img = cv2.normalize(img, None, alpha = 0, beta = 255, norm_type=cv2.NORM_MINMAX)

cv2_imshow(normalized_img)
```

## L2 normalization

```python
normalized_img_l2 = cv2.normalize(img, None, alpha=1, beta=0, norm_type=cv2.NORM_L2, dtype=cv2.CV_32F)

cv2_imshow(normalized_img_l2*255)
```

## Infinity Normalization

```python
normalized_img_inf = cv2.normalize(img, None, alpha = 1, beta = 0, norm_type=cv2.NORM_INF, dtype=cv2.CV_32F)

cv2_imshow(normalized_img_inf*255)
```

## Manual Normalization

```python
normalized_img_man1 = img/255.0
normalized_img_man2 = (img - np.min(img))/(np.max(img)-np.min(img)) *255

cv2_imshow(normalized_img_man1)
cv2_imshow(normalized_img_man2)
```
