```python
from PIL import Image
from PIL.ExifTags import TAGS
import cv2
```

```python
##Advanced MetaData Extraction (I am still not clear on it)

print("\nAdvanced Metadata:")
with Image.open('IMG-20231221-WA0000.jpg') as img:
    exif_data = img._getexif()  # Get EXIF data

    if exif_data:
        for tag_id, value in exif_data.items():
            tag = TAGS.get(tag_id, tag_id)  # Decode the tag name
            print(f"{tag}: {value}")
    else:
        print("No EXIF metadata found.")
```
